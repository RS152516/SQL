# File Handling in Python
File handling refers to the process of performing operations on a file, such as creating, opening, reading, writing and closing it through a programming interface. It involves managing the data flow between the program and the file system on the storage device, ensuring that data is handled safely and efficiently.

# Need for File Handling
Store data permanently, even after the program ends.
Access external files like .txt, .csv, .json, etc.
Process large files efficiently without using much memory.
Automate tasks like reading configs or saving outputs.

# Opening a File
To open a file, we can use open() function, which requires file-path and mode as arguments.

Note: We will use a sample file named geek.txt for all examples in this article. To download click here 

Syntax:

file = open('filename.txt', 'mode')

filename.txt: name (or path) of the file to be opened.
mode: mode in which you want to open the file (read, write, append, etc.).
Note: If you don’t specify the mode, Python uses 'r' (read mode) by default.

# Closing a File
file.close() method closes the file and releases the system resources. If the file was opened in write or append mode, closing ensures that all changes are properly saved.

```python
#opening a file

file = open("C:/Users/yrama/OneDrive/Desktop/Python/doc.txt", "r")

file.close()



```

```python
# Checking File Properties
file = open("C:/Users/yrama/OneDrive/Desktop/Python/doc.txt", "r")
print("Filename:", file.name)
print("Mode:", file.mode)
print("Is Closed?", file.closed)
file.close()
print("Is Closed?", file.closed)
```

# Reading a File
Reading a file can be achieved by file.read() which reads the entire content of the file. After reading, it’s good practice to close the file to free up system resources.

Example: Reading a File in Read Mode (r)

```python
file = open("C:/Users/yrama/OneDrive/Desktop/Python/doc.txt", "r")
content = file.read()
print(content)
file.close()
```

```python
# reading each line 
file = open(r"C:/Users/yrama/OneDrive/Desktop/Python/doc.txt", "r")

for line in file:
    print(line)

file.close()
```

```python
# reading each line
file = open(r"C:/Users/yrama/OneDrive/Desktop/Python/doc.txt", "r")

for line in file:
    input("Press any key to continue...")
    print(line.strip())

file.close()
```

# Writing in a File
Writing to a file is done using the mode "w". This creates a new file if it doesn’t exist, or overwrites the existing file if it does. The write() method is used to add content. After writing, make sure to close the file.

Example: Writing to a file (overwrites if file exists)

```python
file = open("C:/Users/yrama/OneDrive/Desktop/Python/sample.txt", "w")
file.write("Hello Rama\n")
file.write("Welcome to Python\n")
file.write(" i am writing this using pycharm")
file.close()
```

# Append to an existing file
Using "a" mode (append), we can add new data at the end of a file without removing what’s already there.

Example: This writes an extra line to file.txt without removing existing content.

```python
file = open(r"C:/Users/yrama/OneDrive/Desktop/Python/sample.txt", "a")
file.write("Hello Rama\n")
file.write("Welcome to Python\n")
file.close()
```

But important point:

File must already exist.

If file does not exist, r+ gives error.

```python
file = open(r"C:/Users/yrama/OneDrive/Desktop/Python/sample.txt", "r+")
file.write("Hello Rama\n")
file.write("Welcome to Python")
file.close()
```

```python
# seek() is used to move the file pointer/cursor to a specific position in the file.
file = open(r"C:/Users/yrama/OneDrive/Desktop/Python/sample.txt", "r")
print(file.read())
print(file.read())
file.close()     # Because after first read(), pointer is already at the end. There is nothing left to read , so it displays only once 
```

```python
# using seek 
file = open(r"C:/Users/yrama/OneDrive/Desktop/Python/sample.txt", "r")
print(file.read())
file.seek(0)
print(file.read())
file.close()    # now it prints 2 times 
```

# TRY & EXCEPT
Exception Handling allows a program to handle unexpected errors during execution in a controlled way, instead of crashing abruptly. It enables programs to detect errors, manage them properly and continue execution wherever possible.

Handles runtime errors such as invalid input, file not found, division by zero and type mismatches that occur during program execution.
Helps improve program reliability by ensuring the application does not terminate unexpectedly when an error occurs.


```python
# example if we dont use except
num = int(input("Enter number: "))
print(num)
```

```python
# using try and except 
try:
    num = int(input("Enter number: "))
    print(num)

except:
    print("Invalid input. Please enter a number.")
    
```

```python
# using file  
try:
    file = open("C:/Users/yrama/OneDrive/Desktop/Python/dc.txt", "r")
    content = file.read()
    print(content)
    file.close()

except:
    print("File not found. Please check the file name.")
```

# Module 
A module is just a Python file.
    exmaple : calculator.py

Suppose calculator.py has this code:

def add(a, b):
    return a + b

def sub(a, b):
    return a - b

Now we can use this module in another Python file.

```python
import calculator

print(calculator.add(10, 5))
print(calculator.sub(10, 5))
```

# Package 
A package is a folder that contains multiple Python modules.

Example:

my_package/
    __init__.py

# library 
A library is a collection of modules and packages that helps us do some work.

Example libraries:

NumPy    -> numerical calculations
Pandas   -> data analysis
Matplotlib -> charts/graphs
Requests -> API calls
Django   -> web development

```python
import pandas as pd

data = {
    "Name": ["Rama", "Raj", "Ron"],
    "Age": [25, 26, 27]
}

df = pd.DataFrame(data)

print(df)
```
