# Dynamic typing
in Python means you do not need to declare the data type of a variable.
Python automatically understands the data type based on the value you assign.

Example


```python
x = 10
print(type(x))
```

```python
x = "Rama"
print(type(x))
```

# Strong typing 
means Python does not automatically mix different data types without clear conversion.

```python
a = 10
b = "20"

print(a + b)
```

```python
#Operators in Python

a = 5
b = 2
print(a + b)   
print(a - b)   
print(a * b)   
print(a / b)  
print(a % b)  # remainder 
print(a ** b) # a power b 
print(a // b) #floor division
```

```python
#comparison operations

a = 10
b = 5

print(a == b)  
print(a != b)  
print(a < b)   
print(a > b)   
print(a <= b) 
print(a >= b)  
```

```python
#logical operations 

x = True
y = False

print(x and y)  
print(x or y)   
print(not y)   
```

# Branching 
means Python takes a decision and runs code based on a condition.

if 
else
elif
neted if

```python
age = 20

if age >= 18:
    print("You are eligible to vote")
else:
    print("You are not eligible to vote")
```

```python
marks = 45

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```

# Nested if
Use nested if when one condition depends on another condition.
That means: check condition 2 only if condition 1 is True.

```python
username = input("enter user name:")
password = input("enter passwor: ")

if username == "admin":
    if password == "1234":
        print("Login successful")
    else:
        print("Wrong password")
else:
    print("Wrong username")
```
