# Reverse a string using for loop 

```python
text = "RAMA SUNDAR"

reverse_text = ""

for char in text:
    reverse_text = char + reverse_text

print(reverse_text)
```

```python
reverse a string seperated by words
```

```python
text = "RAMA SUNDAR"

words = text.split()  # it conversts [ "RAMA ", "SUNDAR"]
final_result = ""

for word in words:        # it takes each word from words : RAMA & SUNDAR
    reverse_word = ""

    for char in word:                # It first reverses the word RAMA first
        reverse_word = char + reverse_word

    final_result = final_result + reverse_word + " "

print(final_result)
```
