# Encapsulation 
wrapping data and methods together inside a class

# In Python, a class contains:
Variables / attributes  -> data
Methods / functions     -> behavior

When we keep both data and related methods inside one class, it is called encapsulation.

Here, the class Person has data and method together:

```python
class Person:
    def __init__(self, first_name, last_name, age):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age

    def display_details(self):
        print(f"I am {self.first_name} {self.last_name} and my age is {self.age}")


p1 = Person("Rama", "Sundar", 29)
p1.display_details()
```

# Polymorphism 
means , different classes have same method name, different behavior

Here both classes have the same method but different methods



```python
class Dog:
    def sound(self):
        print("Dog barks")


class Cat:
    def sound(self):
        print("Cat meows")


d = Dog()
c = Cat()

d.sound()
c.sound()
```

# Duck Typing in Python

Duck typing means Python does not care about the object’s class/type.

This function does not care whether obj is a Duck or Person.

It only cares that the object has:

speak()

```python
class Duck:
    def speak(self):
        print("Duck says quack")


class Person:
    def speak(self):
        print("Person says hello")


def make_it_speak(obj):
    obj.speak()


d = Duck()
p = Person()

make_it_speak(d)
make_it_speak(p)
```

# Instance Variable
An instance variable belongs to a particular object.
Each object can have its own value.

# Class Variable
A class variable belongs to the class.
It is shared by all objects.


```python
class Person:
    country = "India"   # class variable

    def __init__(self, first_name, last_name, age):
        self.first_name = first_name   # instance variable
        self.last_name = last_name     # instance variable
        self.age = age                 # instance variable

    def display_details(self):
        print(f"I am {self.first_name} {self.last_name}, my age is {self.age}, and I am from {self.country}")


p1 = Person("Rama", "Sundar", 25)
p2 = Person("Raj", "Kumar", 30)

p1.display_details()
p2.display_details()
```

```python
| Class Variable                           | Instance Variable                      |
| ---------------------------------------- | -------------------------------------- |
| Belongs to class                         | Belongs to object                      |
| Shared by all objects                    | Different for each object              |
| Defined inside class but outside methods | Defined inside `__init__` using `self` |
| Accessed using class name or object name | Accessed using object name             |
| Example: `country = "India"`             | Example: `self.age = age`              |
```
