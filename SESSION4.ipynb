# Iterators 
An iterator is an object that gives the next value when we ask.

Python uses two main functions:
iter()
next()

```python
# Iteration using for loop 
names = ["Ram", "Raj", "Ron"]

for name in names:
    print(name)
```

```python
# Iteration using iterators function 
names = ["Ram", "Raj", "Ron"]

it = iter(names)
print(next(it))
print(next(it))
print(next(it))
print(next(it))  # there is no value left. So Python raises StopIteration.
```

# example 
Think of a movie ticket queue:

[Person1, Person2, Person3]

First next()  -> Person1
Second next() -> Person2
Third next()  -> Person3

After that, no one is left.


# Generator 
A generator is used to produce values one at a time, instead of storing all values in memory at once.

Generators are created using the keyword: yield


```python
#
def get_names():
    yield "Rama"
    yield "Raj"
    yield "Ron"

names = get_names()

print(next(names))
print(next(names))
print(next(names))
print(next(names))
```

# mains points for generator
A generator is a special function that uses yield.

yield gives one value and pauses the function.

When next() is called again, it continues from the paused place.

Generators save memory because they do not store all values at once.

| Iterator                                               | Generator                                                    |
| ------------------------------------------------------ | ------------------------------------------------------------ |
| Iterator is an object that gives values one by one     | Generator is a special function that gives values one by one |
| Created using `iter()`                                 | Created using `yield`                                        |
| Uses `next()` to get values                            | Also uses `next()` to get values                             |
| More general concept                                   | Easy way to create an iterator                               |
| Example: `iter([10, 20, 30])`                          | Example: function with `yield`                               |
| Iterator can be created from list, tuple, string, etc. | Generator is created from a function                         |
| Iterator remembers current position                    | Generator also remembers current position                    |


# MAP FUNCTION
map() function in Python applies a given function to each element of an iterable (list, tuple, set, etc.) and returns a map object (iterator). It is a higher-order function used for uniform element-wise transformations, enabling concise and efficient code.
# Syntax
map(function, iterable,..)

Parameters:

function: The function to apply to every element of the iterable.
iterable: One or more iterable objects (list, tuple, etc.) whose elements will be processed.


```python
# Let's start with a simple example of using map() to convert a list of strings into a list of integers.

s = ['1', '2', '3', '4']
res = map(int, s)
print(list(res))
```

```python
# example using maP 
numbers = [1, 2, 3, 4]

def square(num):
    return num * num

result = map(square, numbers)

print(list(result))
```

# Lamda function 
We can use a lambda function instead of a custom function with map() to make code shorter and easier. Let's see how to improve above code for better readability.

```python
# same example using lamda function 
numbers = [1, 2, 3, 4]

result = map(lambda num: num * num, numbers)

print(list(result))
```

```python
# Calculate fahrenheit from celsius using lamda 
celsius = [0, 20, 37, 100]
fahrenheit = map(lambda c: (c * 9/5) + 32, celsius)
print(list(fahrenheit))
```

# REDUCE FUNCTOIN
reduce() function (from the functools module) applies a function cumulatively to the elements of an iterable and returns a single final value. It processes elements step-by-step, combining two elements at a time until only one result remains.


```python
# EXAMPLE FOR REDUCE FUNCTION 

numbers = [10, 45, 22, 90, 5]

biggest = reduce(lambda x, y: x if x > y else y, numbers)

print(biggest)
```

# fiLTER FUNCTION 
filter() function is used to extract elements from an iterable (like a list, tuple or set) that satisfy a given condition. It works by applying a function to each element and keeping only those for which function returns True.



```python
a = [1, 2, 3, 4, 5, 6]
b = filter(lambda x: x % 3 == 0, a)
print(list(b))
```
